```python
import pandas as pd
import numpy as np
from datetime import datetime

class ImpliedVolatilityIndicator:
    def __init__(self, data):
        """
        Initialise la classe avec les données des options.
        """
        self.data = data.copy()
        self.preprocess_data()
    
    def preprocess_data(self):
        """
        Prépare les données pour l'analyse.
        """
        # Convertir les colonnes de dates en format datetime
        self.data['date'] = pd.to_datetime(self.data['date'])
        self.data['expiration'] = pd.to_datetime(self.data['expiration'])
        
        # Calculer les jours jusqu'à l'échéance
        self.data['days_to_expiration'] = (self.data['expiration'] - self.data['date']).dt.days
        
        # Calculer le delta en valeur absolue
        self.data['abs_delta'] = self.data['deltas'].abs()
        
        # Supprimer les observations avec des valeurs manquantes essentielles
        self.data.dropna(subset=['implied volatility', 'deltas', 'days_to_expiration'], inplace=True)
    
    def filter_options(self, delta_target=0.5, delta_tolerance=0.05, dte_target=30, dte_tolerance=5):
        """
        Filtre les options selon le delta et les jours jusqu'à l'échéance.
        """
        # Filtrer les options avec un delta proche de 0.5
        delta_min = delta_target - delta_tolerance
        delta_max = delta_target + delta_tolerance
        self.filtered_data = self.data[
            (self.data['abs_delta'] >= delta_min) & (self.data['abs_delta'] <= delta_max)
        ]
        
        # Filtrer les options avec environ 30 jours à l'échéance
        dte_min = dte_target - dte_tolerance
        dte_max = dte_target + dte_tolerance
        self.filtered_data = self.filtered_data[
            (self.filtered_data['days_to_expiration'] >= dte_min) & (self.filtered_data['days_to_expiration'] <= dte_max)
        ]
    
    def calculate_iv_changes(self):
        """
        Calcule le changement mensuel de la volatilité implicite.
        """
        # Trier les données pour le calcul des différences
        self.filtered_data.sort_values(by=['secid', 'cp_flag', 'date'], inplace=True)
        
        # Calculer le changement de la volatilité implicite
        self.filtered_data['iv_change'] = self.filtered_data.groupby(['secid', 'cp_flag'])['implied volatility'].diff()
        
        # Supprimer les NaN résultant du calcul de différence
        self.filtered_data.dropna(subset=['iv_change'], inplace=True)
    
    def calculate_average_iv(self):
        """
        Calcule la volatilité implicite moyenne des calls et puts ATM pour chaque action.
        """
        # Calculer la volatilité implicite moyenne par secid et date
        self.filtered_data['average_iv'] = self.filtered_data.groupby(['secid', 'date'])['implied volatility'].transform('mean')
    
    def run_analysis(self):
        """
        Exécute toutes les étapes de l'analyse.
        """
        self.filter_options()
        self.calculate_iv_changes()
        self.calculate_average_iv()
        return self.filtered_data
```
